```python
from transformers import pipeline
text_generator=pipeline("text-generation")
text=text_generator("I met him today",truncation=True, num_return_sequences=2)
print(text)
# truncation=True: it means that if the input text is too long for the model's input size, it will be truncated instead of throwing an error
# num_return_sequences: it means that it will generate n sequences for the given input prompt instead of one, jaise yahan 2 hue hain for n=2.
```

```python
qa=pipeline("question-answering")
ques="What is my name?"
hint="My love calls me as Betu"
qa(question=ques, context=hint)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

model="nlptown/bert-base-multilingual-uncased-sentiment" #This is any random model chosen
x=AutoModelForSequenceClassification.from_pretrained(model)
y=AutoTokenizer.from_pretrained(model)

classifier=pipeline("sentiment-analysis", model=x, tokenizer=y)
result=classifier("I love studying sitting right next to him")
print(result)
```

```python
# Application of AutoTokenizer
# Step 1: Convert words to tokens
from transformers import AutoTokenizer

tokenizer=AutoTokenizer.from_pretrained("bert-base-uncased")
text="Hii ! How are you today?"
tokens=tokenizer.tokenize(text)
print(tokens)
```

```python
# Convert tokens into ids
# Step 2: Converting text to vector representation
input_ids=tokenizer.convert_tokens_to_ids(tokens)
print(input_ids)
```

```python
# Step 3: Tokenization- giving attention values to each token. Basically passing through the attention layer of the encoder model
encoded_text=tokenizer(text)
print(encoded_text)
```

```python
 # Step 4: Decoding the text back from the input_ids
 decoded_text=tokenizer.decode(input_ids)
 print(decoded_text)
```
